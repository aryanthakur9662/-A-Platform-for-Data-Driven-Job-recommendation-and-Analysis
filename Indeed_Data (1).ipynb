```python
start = time.time()

job_titles = ['Software Developer', 'Data Scientist', 'UI/UX Designer'] 
locations = ['Hyderabad+Telangana', 'Mumbai+Maharashtra', 'Delhi+Delhi', 'Bangalore+Karnataka']  

job_lst = []
job_description_list_href = []
job_description_list = []
salary_list = []

driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=option)
sleep(randint(2, 6))

for job_ in job_titles:
    for location in locations:
        for i in range(0, max_iter_pgs):
            driver.get(paginaton_url.format(job_, location, i * 10))
            sleep(randint(2, 4))

            job_page = driver.find_element(By.ID, "mosaic-jobResults")
            jobs = job_page.find_elements(By.CLASS_NAME, "job_seen_beacon")

            for jj in jobs:
                try:
                    job_title = WebDriverWait(driver, 10).until(
                        EC.presence_of_element_located((By.CLASS_NAME, "jobTitle"))
                    )
                    job_snippet = WebDriverWait(driver, 10).until(
                        EC.presence_of_element_located((By.CLASS_NAME, "job-snippet"))
                    )

                    job_data = {
                        'job_title': job_title.text,
                        'job_id': job_title.find_element(By.CSS_SELECTOR, "a").get_attribute("id"),
                        'company': jj.find_element(By.CLASS_NAME, "companyName").text,
                        'location': jj.find_element(By.CLASS_NAME, "companyLocation").text,
                        'job_description': job_snippet.text,
                        'salary': None
                    }

                    try:  
                        job_data['salary'] = jj.find_element(By.CLASS_NAME, "salary-snippet-container").text
                    except NoSuchElementException:
                        try:
                            job_data['salary'] = jj.find_element(By.CLASS_NAME, "estimated-salary").text
                        except NoSuchElementException:
                            pass  # Handle the case when salary information is not available

                    job_lst.append(job_data)
                except Exception as e:
                    print(f"An error occurred: {e}")
```

```python
import re

def convert_salary_to_monthly(salary_str):
    salary_str = salary_str.replace('₹', '')  
    salary_pattern = r"([\d,]+) - ([\d,]+) a year"
    match = re.match(salary_pattern, salary_str)
    
    if match:
        lower_salary = int(match.group(1).replace(',', ''))  
        upper_salary = int(match.group(2).replace(',', ''))  
        
        
        average_monthly_salary = (lower_salary + upper_salary) / 24  
        return average_monthly_salary
    
    return None

for job in processed_job_lst:
    if job['salary'] is not None:
        average_salary = convert_salary_to_monthly(job['salary'])
        job['salary'] = average_salary
```

```python
def round_salary_decimals(processed_job_lst):
    for job in processed_job_lst:
        if job['salary'] is not None:
            job['salary'] = round(job['salary']) 
    return processed_job_lst

processed_job_lst = round_salary_decimals(processed_job_lst)
```
